```python
import numpy as np
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Embedding
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences


# Simple RNN for Sequence Prediction

def simple_sequence_prediction():
  """
  Predict the next number in a sample seq.
  """

  # Generate sample sequence
  X = np.array([[i+j for j in range(3)] for i in range(100)])
  y = np.array([i+3 for i in range(100)])


  X = X.reshape((100, 3, 1))

  model = Sequential([
      SimpleRNN(32, input_shape = (3,1)),
      Dense(1)
  ])


  model.compile(optimizer = 'adam', loss='mse')

  model.fit(X, y, epochs = 100, verbose=1)

  test_sequence = np.array([7,8,9]).reshape((1, 3, 1))

  prediction = model.predict(test_sequence)

  print(f" Seq {test_sequence.flatten()} -> Predicted number:{prediction[0][0]:.2f}")


simple_sequence_prediction()
```
